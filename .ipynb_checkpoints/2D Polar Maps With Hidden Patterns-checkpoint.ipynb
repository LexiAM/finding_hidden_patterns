# Generate Data Samples With Hidden Local Patterns

In semiconductor or optics industries production pipelines often perform quality control, continuous improvement or R&D investigations by sampling performance on semiconductor wafers or other substrates over 2D polar maps. Thin films used in fabrication of electronic and opto-electronic devices are sampled for thicknesses, conductivity, optical or other film properties. Resulting 2D maps are used to estimate yields or find unwanted spatial variations.
  
This is a helper notebook for generating 2D polar maps with various spatial trends and hidden patterns: Concave, Convex, Donut-shaped radial patterns and a hidden Smiley Kitty Face.

**Requirements**: Python 3.5+, numpy, pandas, scipy, matplotlib

```python
import numpy as np
import pandas as pd
from scipy.stats import norm

import matplotlib.pyplot as plt
%matplotlib inline
```

## Sampling Map

Below is a function that generates a data frame of polar map coordinates with a given diameter, number of sampling radii, and angular density of points. The output dataframe contains the following columns: **X,  Y, R, Theta**.

```python
# Generate sampling locations
def polar_map(diam, num_radii, angular_density):
    """Returns dataframe containing cartesian coordinates (X, Y) and polar coordinates (R, Theta)
       of a polar sampling map.
    
    Args:
        diam (float): outer diameter of the polar map
        num_radii (int): number of sampling radii, including center and outer edge
        angular_density (int): density of points on each radius.
                               Number of points for a give radius is
                               R0(center): 1 point
                               R1(1st radius): 1 * angular_density points
                               ...
                               Rn(nth radius): n * angular_density points
    
    Returns:
        pm (pandas.DataFrame): data frame df({'X': float, 'Y': float, 'R': float, 'Theta': float})
    
    """
    # initialize X, Y, R, Theta coordinates with center point at 0
    X = np.array([0.])
    Y = np.array([0.])
    R = np.array([0.])
    Theta = np.array([0.])

    # loop through radii and define points at each radius
    for n, radius in enumerate(np.linspace(0, diam / 2, num_radii)):
        if n == 0:
            continue   # skip center point that has already been added at initialization
        
        # define polar angles and ensure not to count 0/360deg twice
        num_points = angular_density * n  # number of sampling points on this radius
        angles = np.radians(np.linspace(0, 360 * (1 - 1 / num_points), num_points))
        
        # generate X, Y, R, Theta coordinates
        X = np.append(X, np.round(radius * np.cos(angles), 2))
        Y = np.append(Y, np.round(radius * np.sin(angles), 2))
        R = np.append(R, np.repeat(radius, num_points))
        Theta = np.append(Theta, np.round(np.degrees(angles), 2))
    
    # store results in data frame
    pm = pd.DataFrame({'X': X, 'Y': Y, 'R': R, 'Theta': Theta})
    
    return pm
```

```python
# Test generate_polar_map(). Map with 13 radii and density_factor=8 should have 625 points
pm = polar_map(diam=294, num_radii=13, angular_density=8)
plt.plot(pm.X, pm.Y, 'o', color='black')
pm.sample(5).sort_index()  # show random slice of the map dataframe
print('Map contains {} sampling points'.format(len(pm.index)))
```

## Sample Patterns  
Below is the code for generating random noise, power radially dependent, donut shaped, and smiley kitty patterns.  
  
The overall pattern is a linear combination of random noise, radial dependence, and hidden patterns:
Pattern = random_noise + radial_dependence + hidden_pattern.  

To represent a realistic use case the default relative amplitudes of the underlying patterns is as follows:  
- Random noise is ~ 0.5% of the overall pattern
- Hidden pattern provides < 2% contribution to the overall pattern  

The relative amplitudes are easily adjusted by changing the **..._amplitude** constants.

```python
### Map
pm = polar_map(294, 13, 8)  # 625-point map

# Relative amplitudes of the contributing factors to overall pattern:
# Overall Pattern = radial dependence + random noise + hidden pattern
radial_dependence_amplitude = 1  # main contributing trend
random_noise_amplitude = 0.005  # ~2% noise level
hidden_pattern_amplitude = -0.02  # 4:1 signal/random noise ratio and <2% overall contribution 

# Random noise pattern
random_noise = np.random.rand(pm.X.shape[0]) * random_noise_amplitude

# Power radial dependence
# Pattern can become concave or convex depending on the sign of radial_dependence
radial = np.power(pm.R / 147, 2.5) * radial_dependence_amplitude

# Bell shape pattern with peak value = radial_dependence_amplitude
nd = norm.pdf(pm.R, pm.R.max() * 0.65, pm.R.max() / 4)
nd_scaled = nd / nd.max()
nd_shifted = nd_scaled - nd_scaled.min()
donut = (nd_shifted / nd_shifted.max()) * radial_dependence_amplitude

# Smiley face pattern
def face_map(row, amplitude):
    if (90 <= row['Y'] <= 100) and ((-75 <= row['X'] <= -35) or (35 <= row['X'] <= 75)): # eyes
        return amplitude
    elif 240 <= row['Theta'] <=300 and row['R'] == 61.25: # smile
        return amplitude
    elif ((49 <= row['R'] <= 110) and 
          ((8 <= row['Theta'] <=16) or (164 <= row['Theta'] <=172)
           or(344 <= row['Theta'] <=352) or (188 <= row['Theta'] <=196))):
        return amplitude
    elif abs(row['X']) < 1 and 5 <= row['Y'] <= 35:  # nose
        return amplitude
    else:
        return 0
smiley_face = pm.apply(face_map, axis=1, args=(hidden_pattern_amplitude,))
```

The radial and countour plots of the resulting patterns are below.

```python
# Plot patterns vs. radius
plt.plot(pm.R, random_noise)
plt.plot(pm.R, radial)
plt.plot(pm.R, donut)
plt.plot(pm.R, smiley_face)
plt.title('Thickness Radial Patterns')
plt.legend(['Random noise', 'Radial power', 'Donut', 'Smiley face'])

# Contour plots of patterns
fig1 = plt.figure(figsize=(20, 5))
fig1.subplots_adjust(top=0.8)

# define grid
ax1 = plt.subplot2grid((1, 4), (0, 0))
ax2 = plt.subplot2grid((1, 4), (0, 1))
ax3 = plt.subplot2grid((1, 4), (0, 2))
ax4 = plt.subplot2grid((1, 4), (0, 3))

#set titles
ax1.set_title('Random Noise', fontweight='bold')
ax2.set_title('Radial Power', fontweight='bold')
ax3.set_title('Donut', fontweight='bold')
ax4.set_title('Smiley Face', fontweight='bold')

# rescale axis
for ax in fig1.axes:
    ax.set_xlim([-150, 150])
    ax.set_ylim([-150, 150])

# generate plots
ax1.tricontourf(pm.X, pm.Y, random_noise)
ax2.tricontourf(pm.X, pm.Y, radial)
ax3.tricontourf(pm.X, pm.Y, donut)
ax4.tricontourf(pm.X, pm.Y, smiley_face)
plt.tight_layout()
```

## Sample Data Sets  
Below we generate sample datasets with concave, convex, and donut radial dependencies with and without hidden pattern. Sample data is stored as a data frame where pattern information is found in column **Value** and spatial coordinates are in columns **X, Y, R,** and **Theta** as above.

```python
data_smiley_kitty = pm.copy().assign(Value=smiley_face)
data_concave = pm.copy().assign(Value=(radial + random_noise + smiley_face))
data_concave_nohidden = pm.copy().assign(Value=(radial + random_noise))
data_convex = pm.copy().assign(Value=(-radial + random_noise + smiley_face))
data_convex_nohidden = pm.copy().assign(Value=(-radial + random_noise))
data_donut = pm.copy().assign(Value=(donut + random_noise + smiley_face))
data_donut_nohidden = pm.copy().assign(Value=(donut + random_noise ))

# Contour plots
fig2 = plt.figure(figsize=(20, 5))
fig2.subplots_adjust(top=0.8)
ax5 = plt.subplot2grid((1, 4), (0, 0))
ax6 = plt.subplot2grid((1, 4), (0, 1))
ax7= plt.subplot2grid((1, 4), (0, 2))
ax8 = plt.subplot2grid((1, 4), (0, 3))
ax5.set_title('Concave', fontweight='bold')
ax6.set_title('Concave Without Hidden Pattern', fontweight='bold')
ax7.set_title('Convex', fontweight='bold')
ax8.set_title('Donut', fontweight='bold')

for ax in [ax5, ax6, ax7, ax8]:
    ax.set_xlim([-150, 150])
    ax.set_ylim([-150, 150])
ax5.tricontourf(data_concave.X, data_concave.Y, data_concave.Value)
ax6.tricontourf(data_concave_nohidden.X, data_concave_nohidden.Y, data_concave_nohidden.Value)
ax7.tricontourf(data_convex.X, data_convex.Y, data_convex.Value)
ax8.tricontourf(data_donut.X, data_donut.Y, data_donut.Value)
plt.tight_layout()
```

```python
# Visualize the data frame
data_donut.sample(5).sort_index()  # show random slice of the sample data set dataframe
```

We export the data sets to csv to be used in our local pattern search analysis.

```python
# Save to csv
data_smiley_kitty.to_csv('data/smiley_kitty.csv', index=False)
data_concave.to_csv('data/concave.csv', index=False)
data_concave_nohidden.to_csv('data/concave_nohidden.csv', index=False)
data_convex.to_csv('data/convex.csv', index=False)
data_convex_nohidden.to_csv('data/convex_nohidden.csv', index=False)
data_donut.to_csv('data/donut.csv', index=False)
data_donut_nohidden.to_csv('data/donut_nohidden.csv', index=False)
```
